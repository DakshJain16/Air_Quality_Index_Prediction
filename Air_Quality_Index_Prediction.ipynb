```python
#missing values are tagged with -200 value instead o NaN
#counting the number of times -200 occurs in the data
air_quality_data.isin([-200]).sum(axis=0)
```

```python
#handling missing values
#convert all -200 to NaN and then replace all NaN with mean of that specific column
air_quality_data = air_quality_data.replace(to_replace = -200, value = np.nan)
```

```python
#checking the number of missing values in the datafrane
air_quality_data.isnull().sum()
```

```python
air_quality_data.tail()
#this shows the actual number of missing values
```

```python
air_quality_data_numeric = air_quality_data.apply(pd.to_numeric, errors='coerce')
mean_values = air_quality_data_numeric.mean()
mean_values
```

```python
#replacing the missing values with mean value
numeric_columns = air_quality_data.select_dtypes(include=['number']).columns
for column in numeric_columns:
    air_quality_data[column] = air_quality_data[column].fillna(air_quality_data[column].mean())
```

```python
#forecasting with FB Prophet Algorithm

#converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'], dayfirst=True)
print(date_info)
```

```python
time_info = air_quality_data['Time']
print(time_info)
```

```python
time_info = time_info.apply(lambda x: x.replace('.', ':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining 2 series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
date_time.head()
```

```python
date_time.shape
```

```python
#combining date and time
date_time['ds'] = date_time['Date'].astype(str)+' '+ date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
#we have to convert ds from object datatype to date time format
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y'] = air_quality_data['RH']
```

```python
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet
```

```python
#training the prophet model

model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='h')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```
